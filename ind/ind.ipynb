# Индивидуальное задание.

## Дана функция затрат $C(x)$, где $x$ - количество произведенных товаров. Необходимо определить стоимость производства товара для заданного диапазона производства и построить график зависимости стоимости от количества произведенных товаров.

### Импортируем библиотеки

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Создаем функцию затрат затрат C(x)

```python
def cost_function(x):
    return 0.5 * x ** 2 + 10 * x + 100
```

### Создаём функцию расчёта общих затрат на производство n товаров

```python
def calculate_total_cost(n):
    x_values = np.linspace(0, n, 100)
    y_values = cost_function(x_values)
    total_cost = np.trapz(y_values, x_values)

    return total_cost
```

```python
# Заданные параметры
n = 10  # Количество произведенных товаров
```

### Расчет общих затрат на производство


```python
total_cost = calculate_total_cost(n)
print(f"Общая стоимость производства {n} товаров: {total_cost}")
```

```python
# Построение графика зависимости стоимости от количества произведенных товаров
x_values = np.linspace(0, n, 100)
y_values = cost_function(x_values)

plt.plot(x_values, y_values)
plt.xlabel('Количество товаров, x')
plt.ylabel('Стоимость производства, C(x)')
plt.title('Зависимость стоимости производства от количества товаров')
plt.grid(True)
plt.show()
```
